## 8.1 Example: Plotting Sine and Cosine Curves *

The imports are there for you below, work through the following steps:

- Make an empty list of `y_values`
- Iterate over a sequence of numbers from 0 to 360 (Google the `range` function)
  - Each iteration, convert the number from degrees to radians *
  - Take the sine of the angle, and append this to the list of `y_values` 
- After the iteration, plot the `y_values`.
- Use the `help` function to access the documentation for `plt.plot`. Use the information on the first page to add 'red plusses' to your plot.
- Include the line `plt.savefig('trig.png')` to save this plot as an image file called `'trig.png'`

> \* radians are an alternate unit for angles.\
> 360 degrees == 2 * pi radians\
> To convert from degrees to radians you multiply by the expression: `2 * pi / 360`

```python
import matplotlib.pyplot as plt
import math

y_values = []

for x_deg in range(361):
    x_rad = x_deg / (2 * math.pi)
    y_values.append(math.sin(x_rad))

plt.plot(y_values, "r+")

plt.savefig("trig_cos.png")
```
